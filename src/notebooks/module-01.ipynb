The dot product is not just a element-wise multiplication. It is a element-wise multiplication followed by a sum. The result generates a single number.

```python
# Implementation of a function that calculates the dot product
def dot_product(u: np.array, v: np.array) -> int:
    assert u.shape[0] == v.shape[0]
    n = u.shape[0]
    result = 0.0
    for i in range(n):
        result = result + u[i] * v[i]

    return result
```

```python
dot_product(a, b)
```

```python
# NumPy has a implemented function that does this calculation
a.dot(b)
```
