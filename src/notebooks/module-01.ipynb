The dot product is not just a element-wise multiplication. It is a element-wise multiplication followed by a sum. The result generates a single number.

```python
# Implementation of a function that calculates the dot product
def dot_product(u: np.array, v: np.array) -> int:
    assert u.shape[0] == v.shape[0]
    n = u.shape[0]
    result = 0.0
    for i in range(n):
        result = result + u[i] * v[i]

    return result
```

```python
dot_product(a, b)
```

```python
# NumPy has a implemented function that does this calculation
a.dot(b)
```

We can also multiply a Matrix by a Vector. We do this by taking the first row of the matrix U and multiply by the vector V, and so on. For this operation to work, we need to make sure that the number of columns in the matrix U is the same as the number of rows in the vector V.

```python
U = np.array([[2, 4, 5, 6], [1, 2, 1, 2], [3, 1, 2, 1]])
```

```python
U
```

```python
# Defining the Matrix Vector multiplication with just Python
def matrix_vector_multiplication(U: np.array, V: np.array) -> np.array:
    assert U.shape[1] == V.shape[0]

    n_rows = U.shape[0]
    result_vector = np.array([dot_product(U[i], V) for i in range(n_rows)])
    return result_vector
```

For a matrix-matrix multiplication, we have 2 matrices, U and V, and the result is a new matrix with the number of rows in U and the number of columns of V. We can only multiply two matrices if the number of columns in the first matrix is equal to the number of rows in the second matrix. In short, if we multiply a matrix A (m x n) by a matrix B (n x p), the resulting matrix Z will be m x p.

```python
# Implementing matrix-matrix multiplication in Python
def matrix_matrix_multiplication(A: np.array, B: np.array) -> np.array:
    assert A.shape[1] == B.shape[0]

    n_rows = A.shape[0]
    n_cols = B.shape[1]

    result = np.zeros((n_rows, n_cols))
    for i in range(n_cols):
        bi = B[:, i]  # Col i of matrix B
        Ubi = matrix_vector_multiplication(A, bi)
        result[:, i] = Ubi
    return result
```

```python
# We can also simplify this function with following code:
def simplified_matrix_matrix_multiplication(A: np.array, B: np.array) -> np.array:
    assert A.shape[1] == B.shape[0]

    return A @ B  # or np.dot(A, B)
```

The __Identity matrix__ is a square matrix where the diagonal is filled with 1 and filled with 0 in every other position.

The __Inverse Matrix__ is a matrix that, when multiplied by its original matrix A, results in the identity matrix. In other words: A * A^-1 = A^-1 * A = I

Important: only square matrices has a inverse matrix.

```python
A = np.array([[4, 2, 3], [4, 5, 6], [7, 8, 9]])
```

```python
inverse = np.linalg.inv(A)
inverse
```

```python
inverse.dot(A)
```
